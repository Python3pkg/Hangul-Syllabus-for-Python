```python
url = 'http://en.wikipedia.org/wiki/Hangul_Syllables'
```

```python
import urllib2
response = urllib2.urlopen(url)
# print response.info()
html = response.read()
response.close()  # best practice to close the file
```

```python
from bs4 import BeautifulSoup
```

```python
content = html.split('class=\"mw-content-ltr\">')[1]
content = content.split('<h2><span class="mw-headline" id="See_also">')[0]
```

```python
soup = BeautifulSoup(content)
# print soup
#print(soup.prettify())
```

```python
soup.td
```

```python
value = soup.findAll('td', style="background:#F8F8F8;font-size:small")
```

```python
value[1].string
```

```python
sset = []
for i in value[1:]:
    if i.string != None:
        t = i.string[2:-1]
        sset.append(t)
```

```python
text = '\u'+sset[0]+'0'
print text.decode('unicode_escape')
# print unicode(text)
# print repr(text).decode("utf-8")
```

```python
for j in sset:
    text = '\u'+j+'0'
    print text.decode('unicode_escape')
```
